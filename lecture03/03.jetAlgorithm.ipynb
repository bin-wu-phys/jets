```python
import numpy as np
```

```python
def pu(pT, y, phi, m):
    if m == 0:
        mT = pT
    else:
        mT = np.sqrt(pT**2 + m**2)
    return np.array([mT*np.cosh(y), pT*np.cos(phi), pT*np.sin(phi),mT*np.sinh(y)])
```

```python
def yphi(xu):
    return np.array([np.sqrt(xu[1]**2 + xu[2]**2), np.arctanh(xu[3]/xu[0]), np.arctan2(xu[2], xu[1])])
```

# $k_T$ algorithm

```python
def dmin(pTls, R):
    min = pTls[:,0].max()**2
    R2 = R**2
    idx = {}
    for i in range(len(pTls)):
        for j in range(i+1,len(pTls)):
            # print((i,j))
            dij = (np.minimum(pTls[i,0], pTls[j,0])**2)*((pTls[i,1]-pTls[j,1])**2 + (pTls[i,2]-pTls[j,2])**2)/R2
            if dij < min:
                min = dij
                idx = (i,j)
    return min, idx
```

```python
def seq(ps, R):
    pTls = np.array([yphi(x) for x in ps])
    diBmin = pTls[:,0].min()**2
    dijmin, idx = dmin(pTls, R)
    if diBmin < dijmin:
        idxB = pTls[:,0].argmin()
        return ps[idxB], np.concatenate([ps[:idxB], ps[idxB + 1:]])
    else:
        return idx, np.concatenate([ps[:idx[0]], ps[idx[0] + 1:idx[1]],
                                    ps[idx[1] + 1:], [ps[idx[0]] + ps[idx[1]]]
                                   ])
    
```

# Anti-$k_T$ algorithm

```python
def dmin(pTls, R):
    min = pTls[:,0].min()**(-2)
    R2 = R**2
    idx = {}
    for i in range(len(pTls)):
        for j in range(i+1,len(pTls)):
            # print((i,j))
            dij = (np.maximum(pTls[i,0], pTls[j,0])**(-2))*((pTls[i,1]-pTls[j,1])**2 + (pTls[i,2]-pTls[j,2])**2)/R2
            if dij < min:
                min = dij
                idx = (i,j)
    return min, idx
```

```python
def seq(ps, R):
    pTls = np.array([yphi(x) for x in ps])
    diBmin = pTls[:,0].max()**(-2)
    dijmin, idx = dmin(pTls, R)
    if diBmin < dijmin:
        idxB = pTls[:,0].argmax()
        return ps[idxB], np.concatenate([ps[:idxB], ps[idxB + 1:]])
    else:
        return idx, np.concatenate([ps[:idx[0]], ps[idx[0] + 1:idx[1]],
                                    ps[idx[1] + 1:], [ps[idx[0]] + ps[idx[1]]]
                                   ])
    
```

# Cambridge/Aachen algorithm

```python
def dmin(pTls, R):
    R2 = R**2
    min = ((pTls[0,1]-pTls[1,1])**2 + (pTls[0,2]-pTls[1,2])**2)/R2
    idx = (0, 1)
    for i in range(len(pTls)):
        for j in range(i+1,len(pTls)):
            # print((i,j))
            dij = ((pTls[i,1]-pTls[j,1])**2 + (pTls[i,2]-pTls[j,2])**2)/R2
            if dij < min:
                min = dij
                idx = (i,j)
    return min, idx
```

```python
def seq(ps, R):
    pTls = np.array([yphi(x) for x in ps])
    dijmin, idx = dmin(pTls, R)
    if dijmin >= 1.0:
        print("All the particles are jets!")
    else:
        print(idx)
        return idx, np.concatenate([ps[:idx[0]], ps[idx[0] + 1:idx[1]],
                                    ps[idx[1] + 1:], [ps[idx[0]] + ps[idx[1]]]
                                   ])
  
```
